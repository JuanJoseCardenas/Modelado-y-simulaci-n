```python
import matplotlib.style
matplotlib.style.use('classic')
import numpy as np
from matplotlib import pyplot as plt
from pylab import *
```

```python
def euler( func, h, tini, tfin, x0, params):
    ti = tini  
    xi = x0
  
    x = list()
    t = list()
  
    while ( ti < tfin ):
    
      x.append (xi)
      t.append (ti)
  
      xi = xi + h*func(ti,xi,params)
      ti = ti + h
    return t, x
```

# Definición de parametros

En la siguiente celda se definen los parametros que se van a usar para hayar los flujos de los valores iniciales dados y también el tiempo en el cual se van a analizar estos.

donde:

h= es el paso a paso que hace la funcion de aproximacion
tini= el tiempo inicial en el cual va a arrancar el analisis
tfin= tiempo final en el cual se va a terminar el analisis
x0= Condicion inicial del problema

```python
h = 0.04
tini = 0
tfin = 3
x0 = -0.5
a=8
```

# Imagen de la soluciíon dada por el metodo de Euler

```python
t1,x1 = euler( segundo, h, tini, tfin, x0, a )

plt.figure()
plt.scatter( t1, x1, c = 'green', label = 'Euler', s = 100)
plt.xlim([0, tfin])
plt.xlabel(r'$t$')
plt.grid()
plt.show()
```

# Imagen que muestra multiples soluciones de diferentes valores iniciales

```python



for i in arange(-1,1,0.5):
    x0=i
    t1,x1 = euler( segundo, h, tini, tfin, x0, a )
    plt.plot( t1, x1, c = 'green')
plt.xlim([0, tfin])
plt.ylim([-1.2, 1.2])
plt.xlabel(r'$t$')
plt.grid()
```
